```python
import numpy as np
```

```python
#creating array from list with type float
a = np.array([[1,2,3],[5,8,7]], dtype ='float')
print ("Array created using passed list:\n",a)
```

```python
b = np.array((1,2,3))
print ("\nArray created using passed tuple:\n",b)
```

```python
#creating a 3*4 array with all zeros
c = np.zeros((3,4))
print("\n",c)
```

```python
#creat a constant value array of complex type
d = np.full((3,3),6,dtype ='complex')
print (d)
```

```python
#create an array with random values
e = np.random.random ((2,2))
print("\nA",e)
```

```python
#creat a sequence of integer
#from 0 to 30 with step 5
f = np.arange(0,30,5)
print(f)
```

```python
g = np.linspace(0,5,10)
print(g)
```

```python
#reshaping 3*4 array to 2*2*3 array
arr = np.array([[1,2,3,4],
               [5,2,5,3],
               [1,3,5,3]])
newarr = arr.reshape(2,2,3)
print("\n",arr)
print("reshaped",newarr)
```
